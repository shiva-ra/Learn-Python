```python
from datetime import datetime
from datetime import timedelta
from datetime import time
from IPython.display import clear_output

def get_date_input(t):

    while True:
        try:

            inp = input(f'Enter {t} Date and Time in this format "YYYY-MM-DD HH:MM" ')
            
            return datetime.strptime(inp,  '%Y-%m-%d %H:%M')

        except ValueError:

            print('Please enter the correct format')
            
def working_hour(t):
    
    while True:
        
        work_start = input(f'Enter working hours {t} time in "HH:MM" ')
        try:
            d = datetime.strptime(work_start,"%H:%M")
            break
        except ValueError:
            print('Please Enter correct time in "HH:MM"')
            
    return time(d.hour, d.minute)

def working_hours_selection():
    
    inp = 0
    
    while inp not in {'1', '2'}:
        inp = input("Enter 1 to continue with 9:00 to 17:00 as Working Hours, else Enter 2 ")
    
    if inp == "1":
        
        working_hours_start = time(9, 0)
        working_hours_end = time(17, 0)
        
    else:
        
        working_hours_start = working_hour('start')
        working_hours_end = working_hour('end')
        
    return working_hours_start, working_hours_end

def weekend_selection():
    
    inp = 0
    
    while inp not in {'1', '2'}:
        inp = input("Enter 1 to continue with Saturday and Sunday as Weekends, else Enter 2 ")
    clear_output()
    if inp == '1':
        
        weekend1 = 5
        weekend2 = 6
    
    else:
        
        weekend1 = weekend2 = 10
        
        while weekend1 == weekend2 or not 0 <= weekend1 <= 6 or not 0 <= weekend2 <= 6:
            try:
                print("Please enter the weekends you wish to choose from the table below")
                print("Days", "     Number to Enter")
                print("Monday", "   0")
                print("Tuesday", "  1")
                print("Wednesday", "2")
                print("Thursday", " 3")
                print("Friday", "   4")
                print("Saturday", " 5")
                print("Sunday", "   6")
                weekend1 = int(input("Enter your First Weekend: "))
                weekend2 = int(input("Enter your Second Weekend: "))
                
            except ValueError:
                
                print("Error, Please Enter Valid Numbers!")                
    
    return weekend1, weekend2
    
# get uptime and downtime input from user
d1 = get_date_input('DownTime')
d2 = get_date_input('UpTime')

# d1 = datetime(2020, 12, 11, 8, 30)
# d2 = datetime(2020, 12, 11, 11, 30)

#validation
while d2 < d1:
    
    print('Uptime should be Greater than Downtime')
    d1 = get_date_input('DownTime')
    d2 = get_date_input('UpTime')
    
# get input for workings hours in a day from user
working_hours_start, working_hours_end = working_hours_selection()
# working_hours_start, working_hours_end = time(8, 30), time(13, 0)

# get input for weekends from user
weekend1, weekend2 = weekend_selection()
# weekend1, weekend2 = 5, 6

# set downtime time from the downtime datetime object
t1 = time(d1.hour, d1.minute)

# set uptime time from the downtime datetime object
t2 = time(d2.hour, d2.minute)

# set downtime date from the downtime datetime object
d1 = datetime(year=d1.year, month=d1.month, day=d1.day)

# set uptime date from the downtime datetime object
d2 = datetime(year=d2.year, month=d2.month, day=d2.day)

current_date = d1
business_hours = 0
non_business_hours = 0

# converting time to timedelta
def tdelta(t):
    return timedelta(hours=t.hour, minutes=t.minute).seconds/3600


business_duration = tdelta(working_hours_end) - tdelta(working_hours_start)
non_business_duration = 24 - business_duration

# Hours effected calculation without time consideration
while current_date <= d2:
    
    if current_date.weekday() == weekend1 or current_date.weekday() == weekend2:
        non_business_hours = non_business_hours + 24
    else:
        business_hours = business_hours + business_duration
        non_business_hours = non_business_hours + non_business_duration
        
    current_date = current_date + timedelta(days=1)
    

# down time edge case: Time Consideration
if t1 <= working_hours_start or d1.weekday() == weekend1 or d1.weekday() == weekend2:

    non_business_hours = non_business_hours - tdelta(t1)

elif t1 > working_hours_start and t1 < working_hours_end:

    non_business_hours = non_business_hours - tdelta(working_hours_start)
    business_hours = business_hours - (tdelta(t1) - tdelta(working_hours_start))

elif t1 >= working_hours_end:

    non_business_hours = non_business_hours - (tdelta(working_hours_start) + tdelta(t1) - tdelta(working_hours_end))
    business_hours = business_hours - business_duration

# up time edge case: Time Consideration
if t2 <= working_hours_start :
    non_business_hours = non_business_hours - (tdelta(working_hours_start) - tdelta(t2) + (24 - tdelta(working_hours_end)))
    business_hours = business_hours - business_duration

elif t2 > working_hours_start and t2 < working_hours_end:
    print(True)
    non_business_hours = non_business_hours - (24-tdelta(working_hours_end))       
    business_hours = business_hours - (tdelta(working_hours_end)-tdelta(t2))

elif t2 >= working_hours_end or d2.weekday() == weekend1 or d2.weekday() == weekend2:
    
    non_business_hours = non_business_hours - (24 - tdelta(t2))
    
# converting time to Hours and Minutes
def convert_time(ts):    
    hrs = int(ts)    
    mns = int((ts - hrs) * 60)    
    s = f'Hours: {hrs}, Minutes: {mns}'    
    return s



print('Business Hours effected : ', convert_time(business_hours))
print(f'Non Business Hours effected : ', convert_time(non_business_hours))
```
